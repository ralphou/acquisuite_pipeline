```python
import numpy as np
import pandas as pd
```

```python
# Clean the CSV files from BuildingOS, concatenate them to be one combined CSV file
data = []
for i in np.arange(0, 16):
    x = str(i).zfill(3)
    csv = pd.read_csv("updateModhopper_csv_files/Obvius_backlog_data_" + x + ".csv")
    for i in np.arange(len(csv.columns)):
        if i==0:
            csv = csv.rename(columns={csv.columns[i]: "Timestamp"})
        else:
            csv = csv.rename(columns={csv.columns[i]: csv[csv.columns[i]].iloc[0]})
    csv = csv.drop([0,1,2], axis=0)
    data.append(csv)
data = pd.concat(data)
data
```

```python
# Want separate dataframes for separate meters, and drop the NaN values
elements = []
for i in np.arange(1, len(csv.columns)):
    x = data[[data.columns[0], data.columns[i]]]
    elements.append(x)
for i in elements:
    i.dropna(inplace=True)
a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p, q, r, s, t, u, v, w, x= elements
```
